#### Count number of tweets by a given Twitter account.

You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url

url_10 = 'https://twitter.com/realdonaldtrump'
````

````python
#your code

html_10 = requests.get(url_10).content
soup_10 = BeautifulSoup(html_10, "lxml")
````

````python
try:
    tweet_box = soup_10.find('li',{'class':'ProfileNav-item ProfileNav-item--tweets is-active'})
    tweets= tweet_box.find('a').find('span',{'class':'ProfileNav-value'})
    print("{} tweets {} number of tweets.".format('realdonaldtrump',tweets.get('data-count')))

except:
    print('Account name not found...')
````

#### Number of followers of a given twitter account

You will need to include a ***try/except block*** in case account/s name not found. 
<br>***Hint:*** the program should count the followers for any provided account

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url_11 = 'https://twitter.com/realdonaldtrump'
````

````python
#your code

html_11 = requests.get(url_11).content
soup_11 = BeautifulSoup(html_11, "lxml")
````

````python
try:
    follow_box = soup_11.find('li',{'class':'ProfileNav-item ProfileNav-item--followers'})
    followers = follow_box.find('a').find('span',{'class':'ProfileNav-value'})
    print("Number of followers: {} ".format(followers.get('data-count')))
    
except:
    print('Account name not found...')
````
